```python
#Nested loops
a=['Red','Blue','Green']
b=['Apple','banana','orange']
for i in a:##outer loop
    for h in b:##inter loop
        print(i,h)
```

```python

for i,h in zip(a,b):
    print(i,h)
```

```python
#Another example of nesting
course = ['Maths','Data Science','Economics','Social Science']
size = [78,90,56,90]#size of batchs A,B,C,D
batch = ('A','B','C','D')
```

```python
#zip function is used for compressing the output
for i,k,l in zip(course,size,batch):
    print(i,k,l)
```

```python
for i in course:##outer loop
    for k in size:##inner loop
        b_loc=size.index(k)
        b=batch[b_loc]
        print(f"Batch {b}of size {k} shall attend {i}")
```
