STRING PARTITION

```python
#PARTITION 
"""
the partition() method searches for a specified string and splits the string into
    -the first elment contains the part before the argument string.
    -the second element contains the argument string .
    - the third element contains the part afterr the argument string.
"""
```

```python
greetings = "hello everyone I am sneha hegde"
```

```python
greetings.partition("I")
```

```python
greetings = "hello everyone I am sneha hegde,how can i help u" 
```

```python
greetings.partition("I")
```
